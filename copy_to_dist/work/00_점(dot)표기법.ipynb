# 🔵 00. 점(dot, .) 표기법 — Python에서 가장 많이 쓰는 문법

pandas 코드를 보면 `.`(점)이 곳곳에 등장합니다.

```python
import pandas as pd          # 패키지 import

df = pd.DataFrame(데이터)    # pd.DataFrame → pd의 DataFrame
df.head()                     # df.head()    → df의 head 메서드
df.shape                      # df.shape     → df의 shape 속성
```

이 세 줄에서 `.`은 모두 **같은 문법**이지만 맥락이 다릅니다.  
이 노트북에서 점(.)의 두 가지 역할을 완전히 이해합니다.

## 학습 목표
| 장 | 주제 | 핵심 |
|----|------|------|
| 1장 | 점(.)이란? | "~의" 접근 연산자 |
| 2장 | 패키지에서의 점 | `math.pi`, `pd.DataFrame`, `np.random.randint()` |
| 3장 | 클래스 내부의 점 | `self.속성`, `self.메서드()` |
| 4장 | 인스턴스에서의 점 | `df.shape` vs `df.head()` (속성 vs 메서드) |
| 5장 | 메서드 체이닝 | `df.groupby().sum().sort_values().head()` |

> 이 노트북을 먼저 보고 **01_필수 → 02_핵심 → 03_중요 → 04_심화** 순서로 진행하세요.

## 1장. 점(.)이란 무엇인가?

점(`.`)은 한국어로 번역하면 **"~의"** 또는 **"~안에 있는"** 입니다.

```
pd.DataFrame   →  pd 의  DataFrame
df.head()      →  df 의  head 메서드
self.이름      →  self 의  이름 속성
math.pi        →  math 모듈 안에 있는 pi 상수
```

Python에서 `.`이 등장하는 두 가지 상황:

```
상황 1. 패키지/모듈 접근
   import math
   math.sqrt(4)    ← math 모듈 안의 sqrt 함수

상황 2. 객체(클래스 인스턴스)의 멤버 접근
   df.head()       ← df 객체의 head 메서드
   df.shape        ← df 객체의 shape 속성
```

두 상황 모두 **"왼쪽.오른쪽"** 구조이고, 의미는 **"왼쪽 안에 있는 오른쪽"**입니다.

````python
# =============================================
# 1-1. 점(.)의 두 가지 역할 미리보기
# =============================================

# 설명: 이 코드 한 블록에서 점(.)이 몇 번 나오는지 세어보세요!

import math       # 모듈 import
import pandas as pd

# ① 패키지/모듈에서의 점
파이값 = math.pi                # math 모듈 안의 pi 상수
루트4  = math.sqrt(4)           # math 모듈 안의 sqrt 함수

데이터 = {'이름': ['Alice', 'Bob'], '점수': [90, 85]}
df = pd.DataFrame(데이터)       # pd 패키지 안의 DataFrame 클래스

# ② 객체(인스턴스)에서의 점
행수, 열수 = df.shape           # df 객체의 shape 속성 (괄호 없음)
첫행 = df.head(1)               # df 객체의 head 메서드 (괄호 있음)

print('=== ① 패키지에서의 점 ===')
print(f'math.pi   = {파이값}')
print(f'math.sqrt(4) = {루트4}')

print('\n=== ② 객체에서의 점 ===')
print(f'df.shape  = {df.shape}  (행수={행수}, 열수={열수})')
print(f'df.head(1):')
print(첫행)

print('\n--- 핵심 ---')
print('점(.)의 의미: "왼쪽 안에 있는 오른쪽을 꺼내주세요"')
````

## 2장. 패키지/모듈에서의 점 — `import`와 함께 사용

Python 패키지(라이브러리)는 **폴더 구조**로 되어 있습니다.  
`import`로 패키지를 불러오면, `.`으로 그 안의 함수·상수·클래스에 접근합니다.

```
폴더 비유:
  pandas/
  ├── DataFrame  ← pd.DataFrame
  ├── Series     ← pd.Series
  ├── read_csv   ← pd.read_csv
  └── ...

  numpy/
  ├── array      ← np.array
  ├── random/    ← 서브 패키지
  │   ├── randint  ← np.random.randint
  │   └── seed     ← np.random.seed
  └── ...
```

````python
# =============================================
# 2-1. math 모듈 — 가장 간단한 예시
# =============================================

# 설명: import하면 모듈을 변수처럼 사용 가능
#       모듈.함수() 또는 모듈.상수 형태로 접근

import math  # math 모듈 전체를 가져옴

# 상수 접근 (괄호 없음 → 값 자체)
print('=== math 모듈의 상수 ===')
print(f'math.pi     = {math.pi}')      # 원주율 π
print(f'math.e      = {math.e}')       # 자연상수 e
print(f'math.inf    = {math.inf}')     # 무한대

# 함수 접근 (괄호 있음 → 실행)
print('\n=== math 모듈의 함수 ===')
print(f'math.sqrt(16)  = {math.sqrt(16)}')   # 제곱근
print(f'math.floor(3.7)= {math.floor(3.7)}') # 내림
print(f'math.ceil(3.2) = {math.ceil(3.2)}')  # 올림
print(f'math.log(100, 10)= {math.log(100, 10)}')  # 로그

# 핵심 정리
print('\n--- 패턴 정리 ---')
print('  math.pi      → math 모듈의 pi 상수 (값)')
print('  math.sqrt(x) → math 모듈의 sqrt 함수 (실행)')
````

````python
# =============================================
# 2-2. import 방식에 따른 점 표기법 변화
# =============================================

# 설명: import 방식에 따라 점을 쓰는 방식이 달라짐
# 주의: 'from ... import ...' 쓰면 점 없이 바로 사용 가능
#       하지만 어디서 온 함수인지 불명확해져 코드 가독성이 떨어질 수 있음

# 방법 1: import 전체 모듈 → 항상 모듈명.함수() 형태
import math
결과1 = math.sqrt(25)      # math.sqrt → math 모듈의 sqrt
print(f'방법1 (import math): math.sqrt(25) = {결과1}')

# 방법 2: import 별칭 → 짧은 이름으로 접근
import numpy as np          # numpy를 np라는 이름으로
결과2 = np.sqrt(25)        # np.sqrt → numpy 모듈의 sqrt
print(f'방법2 (import numpy as np): np.sqrt(25) = {결과2}')

# 방법 3: from ... import → 점 없이 직접 사용
from math import sqrt, pi   # sqrt와 pi만 직접 가져옴
결과3 = sqrt(25)            # 점 없이 바로 sqrt 사용
print(f'방법3 (from math import sqrt): sqrt(25) = {결과3}')

print('\n=== pandas와 numpy 관례 ===')
print('  import pandas as pd    → pd.XXX 형태로 항상 사용')
print('  import numpy as np     → np.XXX 형태로 항상 사용')
print('  (관례: 모든 pandas 코드는 pd., 모든 numpy 코드는 np. 로 시작)')
````

````python
# =============================================
# 2-3. pandas 패키지의 점 구조
# =============================================

# 설명: pandas 패키지 안에 무엇이 있는지 직접 확인
# pandas 연결: pd. 뒤에 무엇을 붙일 수 있는지 이해

import pandas as pd

print('=== pd. 뒤에 올 수 있는 것들 ===')

# 클래스 (대문자로 시작)
print('\n[클래스] → pd.클래스명(인수)로 인스턴스 생성')
print('  pd.DataFrame  :', type(pd.DataFrame))   # 클래스 자체 출력
print('  pd.Series     :', type(pd.Series))
print('  pd.Index      :', type(pd.Index))

# 함수 (소문자로 시작)
print('\n[함수] → pd.함수명(인수)로 바로 호출')
print('  pd.read_csv   :', type(pd.read_csv))    # 함수 객체 출력
print('  pd.to_numeric :', type(pd.to_numeric))
print('  pd.concat     :', type(pd.concat))
print('  pd.merge      :', type(pd.merge))

# 실제 사용 예시
print('\n=== 실제 사용 예시 ===')
# pd.DataFrame → 클래스 → () 로 인스턴스 생성
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print('pd.DataFrame({...}) 결과:')
print(df)

# pd.concat → 함수 → 직접 호출
df2 = pd.concat([df, df], ignore_index=True)
print(f'\npd.concat([df, df]) → {len(df2)}행')
````

````python
# =============================================
# 2-4. 중첩 점 표기법 — 서브 패키지 접근
# =============================================

# 설명: 점을 연속으로 써서 패키지 → 서브패키지 → 함수 깊이 접근
# 예시: np.random.randint()
#        ↑       ↑       ↑
#       numpy  random  randint 함수
#       패키지 서브모듈

import numpy as np

print('=== numpy의 서브 패키지 구조 ===')

# np.random 서브모듈
print('\n[np.random 서브모듈]')
np.random.seed(42)  # 재현성을 위해 시드 고정
print(f'  np.random.randint(1, 10, 5) = {np.random.randint(1, 10, 5)}')
print(f'  np.random.uniform(0, 1, 3)  = {np.random.uniform(0, 1, 3).round(3)}')
print(f'  np.random.choice(["A","B","C"], 5) = {np.random.choice(["A","B","C"], 5)}')

# np.linalg 서브모듈
print('\n[np.linalg 서브모듈 - 선형대수]')
행렬 = np.array([[1, 2], [3, 4]])
print(f'  np.linalg.det([[1,2],[3,4]]) = {np.linalg.det(행렬):.1f}')  # 행렬식

# 읽는 방법
print('\n=== 점 표기법 읽는 방법 ===')
print('  np.random.randint(1, 10, 5)')
print('  → "numpy 패키지 안의 random 서브모듈 안의 randint 함수 실행"')
print()
print('  pd.read_csv("파일.csv")')
print('  → "pandas 패키지 안의 read_csv 함수 실행"')
````

## 3장. 클래스 내부에서의 점 — `self.`

클래스 안에서 `self.`는 **"이 인스턴스의"**를 의미합니다.  
`self.속성명`으로 데이터를 저장하고, `self.메서드명()`으로 같은 클래스의 다른 메서드를 호출합니다.

```python
class 상품:
    def __init__(self, 이름, 가격):
        self.이름 = 이름      ← "이 인스턴스의 이름" 저장
        self.가격 = 가격      ← "이 인스턴스의 가격" 저장

    def 할인가(self):
        return self.가격 * 0.9  ← "이 인스턴스의 가격"에 접근
```

→ `self`가 없으면 어떤 인스턴스의 데이터인지 알 수 없음!

````python
# =============================================
# 3-1. self 없이는 어떤 일이 생기나?
# =============================================

# 설명: self가 필요한 이유를 직접 느껴보는 예제
# 동일한 클래스로 만든 두 인스턴스가 서로 독립적인 데이터를 가짐

class 상품:
    """상품 정보를 담는 클래스"""
    
    def __init__(self, 이름, 가격):
        # self.이름 = "이 인스턴스(객체)의 이름 속성"
        # self.가격 = "이 인스턴스(객체)의 가격 속성"
        self.이름 = 이름   # 각 인스턴스가 독립적으로 저장
        self.가격 = 가격
    
    def 정보_출력(self):
        # self.이름 → 호출된 인스턴스의 이름을 가져옴
        print(f'상품: {self.이름}, 가격: {self.가격:,}원')
    
    def 할인가(self, 할인율=0.1):
        # self.가격 → 호출된 인스턴스의 가격을 사용
        return int(self.가격 * (1 - 할인율))


# 같은 클래스(설계도)로 다른 인스턴스(제품) 만들기
노트북 = 상품('노트북', 1200000)  # 노트북 인스턴스 생성
마우스 = 상품('마우스', 35000)    # 마우스 인스턴스 생성

print('=== 두 인스턴스는 독립적인 데이터를 가짐 ===')
노트북.정보_출력()   # 노트북의 self.이름, self.가격 사용
마우스.정보_출력()   # 마우스의 self.이름, self.가격 사용

print('\n=== 각 인스턴스에서 독립적으로 메서드 호출 ===')
print(f'노트북 10% 할인가: {노트북.할인가():,}원')  # 노트북.가격 * 0.9
print(f'마우스 10% 할인가: {마우스.할인가():,}원')  # 마우스.가격 * 0.9

print('\n--- self가 하는 일 ---')
print('노트북.정보_출력() 호출 시: self = 노트북 인스턴스')
print('마우스.정보_출력() 호출 시: self = 마우스 인스턴스')
print('→ self 덕분에 같은 메서드가 각자의 데이터를 사용')
````

````python
# =============================================
# 3-2. self.속성 vs self.메서드()
# =============================================

# 설명: 클래스 내부의 점(.)도 속성과 메서드로 나뉨
# 규칙: self.속성  → 데이터 저장·조회 (괄호 없음)
#        self.메서드() → 기능 실행 (괄호 있음)

class 판매_상품:
    """self.속성과 self.메서드()를 모두 보여주는 예시 클래스"""
    
    def __init__(self, 이름, 가격, 수량):
        # ★ self.속성 = 값 → 데이터 저장
        self.이름 = 이름             # 속성: 상품 이름
        self.가격 = 가격             # 속성: 단가
        self.수량 = 수량             # 속성: 판매 수량
        self.매출 = 가격 * 수량      # 속성: 계산된 값도 저장 가능
    
    def 요약(self):
        """★ self.메서드() → 다른 메서드 호출"""
        print(f'[요약] ', end='')
        self.정보_출력()   # ← self.메서드() 로 같은 클래스의 메서드 호출!
        print(f'       매출: {self.매출:,}원')  # self.속성 으로 데이터 접근
    
    def 정보_출력(self):
        # self.이름, self.가격 → 속성 접근
        print(f'{self.이름} | {self.가격:,}원 × {self.수량}개')
    
    def 고수익_여부(self, 기준=500000):
        # self.매출 → 속성 접근하여 조건 판단
        return self.매출 >= 기준


노트북 = 판매_상품('노트북', 1200000, 5)
마우스 = 판매_상품('마우스', 35000, 50)

print('=== self.속성 직접 접근 ===')
print(f'노트북.이름  = {노트북.이름}')     # 속성 접근 (괄호 없음)
print(f'노트북.가격  = {노트북.가격:,}원')
print(f'노트북.매출  = {노트북.매출:,}원')

print('\n=== self.메서드() 호출 ===')
노트북.요약()   # 메서드 호출 (괄호 있음) → 내부에서 self.정보_출력() 호출
마우스.요약()

print('\n=== self.속성을 이용한 판단 ===')
print(f'노트북 고수익? {노트북.고수익_여부()}')
print(f'마우스 고수익? {마우스.고수익_여부()}')
````

````python
# =============================================
# 3-3. self 없이 쓰면 어떻게 되나? (오류 시연)
# =============================================

# 설명: self.속성 대신 일반 변수로 쓰면 어떤 문제가 생기는지
# 주의: 이 셀은 의도적으로 잘못된 코드를 보여줍니다

# 잘못된 클래스: self 없이 지역변수 사용
class 잘못된_상품:
    def __init__(self, 이름, 가격):
        이름_지역변수 = 이름   # self 없음 → __init__ 안에서만 살다가 사라짐!
        가격_지역변수 = 가격   # self 없음 → 메서드 끝나면 소멸
    
    def 정보_출력(self):
        # 문제: 이름_지역변수와 가격_지역변수는 __init__에서 사라짐
        # 이 메서드에서는 접근 불가!
        try:
            print(이름_지역변수)  # NameError 발생!
        except NameError as e:
            print(f'오류 발생: {e}')
            print('→ self 없이 저장한 변수는 다른 메서드에서 접근 불가!')

# 올바른 클래스: self 사용
class 올바른_상품:
    def __init__(self, 이름, 가격):
        self.이름 = 이름   # self → 인스턴스에 저장 → 어디서든 접근 가능
        self.가격 = 가격
    
    def 정보_출력(self):
        print(f'{self.이름}: {self.가격:,}원')  # self.이름 → 정상 접근!

print('=== 잘못된 클래스 (self 없음) ===')
잘못된_객체 = 잘못된_상품('노트북', 1200000)
잘못된_객체.정보_출력()

print('\n=== 올바른 클래스 (self 있음) ===')
올바른_객체 = 올바른_상품('노트북', 1200000)
올바른_객체.정보_출력()

print('\n--- 결론 ---')
print('self.속성 → 인스턴스 전체에서 공유되는 "인스턴스 변수"')
print('일반 변수 → 해당 메서드 안에서만 살아있는 "지역 변수"')
````

## 4장. 인스턴스(객체)에서의 점 — 속성 vs 메서드

클래스로 만든 인스턴스(객체)에 `.`을 붙이면 두 가지 중 하나에 접근합니다:

```
객체.속성    → 객체가 가진 데이터 (괄호 없음)
객체.메서드() → 객체가 가진 기능 실행 (괄호 있음)
```

**pandas DataFrame도 클래스의 인스턴스**이므로 완전히 같은 원리입니다:

```python
df = pd.DataFrame(...)  # DataFrame 클래스의 인스턴스 생성

df.shape     # 속성: (행수, 열수) 튜플 → 괄호 없음
df.columns   # 속성: 컬럼명 리스트 → 괄호 없음
df.dtypes    # 속성: 컬럼별 타입 → 괄호 없음

df.head()    # 메서드: 처음 5행 반환 → 괄호 있음
df.info()    # 메서드: 요약 정보 출력 → 괄호 있음
df.describe()# 메서드: 통계 요약 → 괄호 있음
```

````python
# =============================================
# 4-1. 커스텀 클래스와 DataFrame 나란히 비교
# =============================================

# 설명: 직접 만든 클래스와 pandas DataFrame의 점 표기법을 1:1 비교
# 핵심: 두 곳에서 사용되는 .이 완전히 같은 문법!

import pandas as pd

# ── 커스텀 클래스 ──
class 데이터_컨테이너:
    """DataFrame처럼 속성과 메서드를 가진 클래스"""
    
    def __init__(self, 데이터_딕셔너리):
        self.데이터 = 데이터_딕셔너리
        self.컬럼목록 = list(데이터_딕셔너리.keys())     # 속성
        self.행수 = len(list(데이터_딕셔너리.values())[0])  # 속성
    
    def 미리보기(self, n=2):    # 메서드 (괄호 있음)
        for 컬럼 in self.컬럼목록:
            print(f'  {컬럼}: {self.데이터[컬럼][:n]}')

# 인스턴스 생성
판매_데이터 = {
    '상품명': ['노트북', '마우스', '키보드'],
    '가격':   [1200000, 35000, 75000],
    '수량':   [5, 50, 30],
}
내_컨테이너 = 데이터_컨테이너(판매_데이터)   # 커스텀 인스턴스
df = pd.DataFrame(판매_데이터)               # DataFrame 인스턴스

print('=' * 50)
print('  커스텀 클래스          pandas DataFrame')
print('=' * 50)

print(f'\n[속성 - 괄호 없음]')
print(f'  내_컨테이너.컬럼목록 = {내_컨테이너.컬럼목록}')
print(f'  df.columns           = {list(df.columns)}')

print(f'\n  내_컨테이너.행수     = {내_컨테이너.행수}')
print(f'  df.shape             = {df.shape}')

print(f'\n[메서드 - 괄호 있음]')
print('  내_컨테이너.미리보기(2):')
내_컨테이너.미리보기(2)
print('  df.head(2):')
print(df.head(2))
````

````python
# =============================================
# 4-2. 속성 vs 메서드 구분 연습
# =============================================

# 설명: 괄호 유무로 속성/메서드를 구분하는 연습
# 핵심 규칙:
#   속성  → 값을 저장한 것 → 이미 계산되어 있음 → 괄호 없음
#   메서드 → 실행하는 것   → 호출해야 결과 나옴 → 괄호 있음

import pandas as pd

df = pd.DataFrame({
    '상품명':   ['노트북', '마우스', '키보드', '모니터'],
    '가격':     [1200000, 35000, 75000, 450000],
    '수량':     [5, 50, 30, 8],
    '카테고리': ['IT기기', 'IT기기', 'IT기기', 'IT기기'],
})
df['매출'] = df['가격'] * df['수량']

print('=' * 55)
print('  DataFrame 속성 (속성 = 괄호 없음, 이미 계산된 값)')
print('=' * 55)

print(f'  df.shape       = {df.shape}       ← (행수, 열수) 튜플')
print(f'  df.ndim        = {df.ndim}            ← 차원수 (DataFrame=2)')
print(f'  df.size        = {df.size}           ← 전체 원소 수 (행×열)')
print(f'  df.columns     = {list(df.columns)}')
print(f'  df.index       = {list(df.index)}')
print(f'  df.dtypes:')
for 컬럼, 타입 in df.dtypes.items():
    print(f'    {컬럼:<8} {타입}')

print()
print('=' * 55)
print('  DataFrame 메서드 (메서드 = 괄호 있음, 실행해야 결과)')
print('=' * 55)
print('  df.head(2):')
print(df.head(2))
print('  df.describe():')
print(df.describe().round(0))
````

````python
# =============================================
# 4-3. 속성과 메서드를 혼동하면 생기는 오류
# =============================================

# 설명: 실수로 괄호를 잘못 붙이면 어떤 오류가 나는지 시연
# 주의: 의도적인 오류 예시입니다

import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})

# 실수 1: 속성에 괄호를 붙이면?
print('=== 실수 1: 속성에 괄호 붙이기 ===')
try:
    결과 = df.shape()     # shape는 속성! 괄호 불필요
except TypeError as e:
    print(f'오류: {e}')
    print('→ df.shape는 속성 (이미 값이 있음) → 괄호 빼기: df.shape')

# 올바른 방법
print(f'올바른 방법: df.shape = {df.shape}')

# 실수 2: 메서드에 괄호를 빼면?
print('\n=== 실수 2: 메서드에 괄호 빼기 ===')
결과 = df.head    # 괄호 없음 → 메서드 '객체'가 반환됨 (실행 X)
print(f'df.head (괄호 없음) → {결과}')  # 함수 객체 출력
print('→ 메서드는 괄호를 붙여야 실행됨: df.head()')
print(f'\n올바른 방법: df.head():')
print(df.head())

print('\n--- 암기 팁 ---')
print('  괄호 없음 = 속성 = 이미 저장된 값')
print('  괄호 있음 = 메서드 = 지금 실행해서 결과를 얻음')
````

## 5장. 메서드 체이닝 — 점을 연속으로 사용

pandas 코드에서 점(`.`)을 여러 번 연결해서 쓰는 것을 **메서드 체이닝(Method Chaining)**이라 합니다.

```python
df.groupby('카테고리')['매출'].sum().sort_values(ascending=False).head(3)
   ↑                            ↑      ↑                              ↑
   1단계: 그룹별 묶기         2단계  3단계: 정렬               4단계: 상위 3개
```

**왜 이게 가능한가?** → 각 메서드가 새로운 **객체를 반환**하기 때문!  
반환된 객체에 다시 `.`을 붙여 다음 메서드를 호출합니다.

````python
# =============================================
# 5-1. 메서드 체이닝이 가능한 이유
# =============================================

# 설명: 메서드가 객체를 반환하면 바로 다음 메서드를 연결 가능
# 같은 결과를 단계별로 vs 체이닝으로 비교

import pandas as pd

df = pd.DataFrame({
    '상품명':   ['노트북', '마우스', '키보드', '모니터', '헤드셋', '웹캠'],
    '카테고리': ['IT기기', 'IT기기', 'IT기기', 'IT기기', '음향기기', '영상기기'],
    '가격':     [1200000, 35000, 75000, 450000, 89000, 120000],
    '수량':     [5, 50, 30, 8, 20, 12],
})
df['매출'] = df['가격'] * df['수량']

print('=== 단계별 코드 (중간 변수 사용) ===')
단계1 = df.groupby('카테고리')           # 카테고리로 그룹화 → GroupBy 객체
단계2 = 단계1['매출']                    # 매출 컬럼 선택 → SeriesGroupBy
단계3 = 단계2.sum()                      # 합산 → Series
단계4 = 단계3.sort_values(ascending=False)  # 정렬 → Series
결과  = 단계4.head(3)                    # 상위 3개 → Series
print(결과)

print('\n=== 메서드 체이닝 (연속 점 표기법) — 결과 동일 ===')
결과_체이닝 = (
    df
    .groupby('카테고리')            # 1단계: 그룹화
    ['매출']                        # 2단계: 컬럼 선택
    .sum()                          # 3단계: 합산
    .sort_values(ascending=False)   # 4단계: 내림차순 정렬
    .head(3)                        # 5단계: 상위 3개
)
print(결과_체이닝)

print(f'\n결과 동일?: {결과.equals(결과_체이닝)}')
````

````python
# =============================================
# 5-2. return self로 직접 체이닝 구현
# =============================================

# 설명: 커스텀 클래스에서도 체이닝이 가능하게 만드는 방법
#       핵심: 각 메서드가 self를 반환 → 다음 메서드 연결 가능
# pandas 연결: 04_심화.ipynb의 파이프라인 클래스와 동일한 패턴

class 문자열_처리기:
    """문자열 변환을 체이닝으로 수행하는 클래스"""
    
    def __init__(self, 텍스트):
        self.텍스트 = 텍스트
    
    def 소문자(self):
        self.텍스트 = self.텍스트.lower()
        return self   # ← 핵심! self를 반환하면 다음 메서드 연결 가능
    
    def 공백제거(self):
        self.텍스트 = self.텍스트.strip()
        return self   # ← self 반환
    
    def 특수문자_제거(self):
        import re
        self.텍스트 = re.sub(r'[^가-힣a-zA-Z0-9 ]', '', self.텍스트)
        return self   # ← self 반환
    
    def 결과(self):
        return self.텍스트  # 최종 결과 반환


# 단계별 방법
처리기1 = 문자열_처리기('  Hello, 안녕하세요! 2024  ')
처리기1.소문자()
처리기1.공백제거()
처리기1.특수문자_제거()
print('단계별 결과:', 처리기1.결과())

# 체이닝 방법 (return self 덕분에 가능)
결과 = (
    문자열_처리기('  Hello, 안녕하세요! 2024  ')
    .소문자()           # ← 메서드 실행 후 self 반환
    .공백제거()         # ← 반환된 self에 다시 . 접근
    .특수문자_제거()    # ← 또다시 . 접근
    .결과()             # ← 최종 결과 반환
)
print('체이닝 결과:', 결과)

print('\n--- 체이닝 원리 ---')
print('  메서드가 self(자기 자신)를 반환')
print('  → 반환된 self에 바로 .다음메서드() 호출 가능')
print('  → 마치 한 줄로 이어서 쓰는 것처럼 보임')
````

````python
# =============================================
# 5-3. pandas pipe() — 체이닝의 확장
# =============================================

# 설명: pipe()는 외부 함수를 체이닝에 끼워넣을 수 있는 메서드
# pandas 연결: 04_심화.ipynb에서 자세히 다룸
# 여기서는 점(.)의 관점에서 pipe()를 이해

import pandas as pd

df = pd.DataFrame({
    '상품명':   ['노트북', '마우스', '키보드', '모니터', '헤드셋', '웹캠'],
    '가격':     [1200000, 35000, 75000, 450000, 89000, 120000],
    '수량':     [5, 50, 30, 8, 20, 12],
})

# 전처리 함수들 (외부 함수)
def 매출_추가(데이터프레임):
    df_새 = 데이터프레임.copy()
    df_새['매출'] = df_새['가격'] * df_새['수량']
    print('  → 매출 컬럼 추가')
    return df_새

def 고가_필터(데이터프레임, 기준=100000):
    결과 = 데이터프레임[데이터프레임['가격'] >= 기준]
    print(f'  → 가격 {기준:,}원 이상 필터링 ({len(결과)}개 남음)')
    return 결과

def 정렬(데이터프레임, 기준='매출'):
    결과 = 데이터프레임.sort_values(기준, ascending=False)
    print(f'  → {기준} 기준 내림차순 정렬')
    return 결과

# pipe()로 외부 함수를 체이닝에 연결
print('=== df.pipe()로 함수 체이닝 ===')
결과 = (
    df
    .pipe(매출_추가)              # 외부 함수를 체이닝에 끼워넣기
    .pipe(고가_필터, 기준=100000) # 인수도 전달 가능
    .pipe(정렬, 기준='매출')
)
print('\n최종 결과:')
print(결과[['상품명', '가격', '매출']])

print('\n--- pipe()의 역할 ---')
print('  df.pipe(함수) = 함수(df) 와 동일')
print('  체이닝 흐름을 끊지 않고 외부 함수 삽입 가능')
````

## 6장. 전체 정리 — 점(.)의 모든 경우

````python
# =============================================
# 6-1. 점(.) 표기법 총정리 코드
# =============================================

# 설명: 하나의 코드 블록에서 점의 모든 역할을 확인

import math          # 패키지 import
import pandas as pd  # 별칭으로 import
import numpy as np

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 상황 1: 패키지.함수() / 패키지.상수
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
print('[1] 패키지에서의 점')
print(f'  math.pi              = {math.pi:.4f}')        # 패키지.상수
print(f'  math.sqrt(9)         = {math.sqrt(9)}')        # 패키지.함수()
print(f'  np.random.randint(1,10,3) = {np.random.randint(1,10,3)}')  # 패키지.서브모듈.함수()

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 상황 2: 클래스 내부의 self.
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
print('\n[2] 클래스 내부의 self.')

class 점_예시_클래스:
    def __init__(self, 값):
        self.값 = 값             # self.속성 = 데이터 저장
        self.제곱 = 값 ** 2
    
    def 출력(self):
        print(f'  self.값={self.값}, self.제곱={self.제곱}')  # self.속성 접근
        self.세제곱_출력()  # self.메서드() 호출
    
    def 세제곱_출력(self):
        print(f'  세제곱={self.값 ** 3}')

obj = 점_예시_클래스(3)
obj.출력()

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 상황 3: 인스턴스.속성 / 인스턴스.메서드()
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
print('\n[3] 인스턴스에서의 점')
df = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})
print(f'  df.shape   = {df.shape}  ← 속성 (괄호 없음)')
print(f'  df.columns = {list(df.columns)}  ← 속성 (괄호 없음)')
print(f'  df.head(1):  ← 메서드 (괄호 있음)')
print(df.head(1))

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# 상황 4: 메서드 체이닝
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
print('\n[4] 메서드 체이닝 (점 연속 사용)')
결과 = df['A'].apply(lambda x: x*10).sort_values(ascending=False).head(2)
print(f'  df["A"].apply().sort_values().head() = {list(결과)}')
````

## 최종 정리표

| 상황 | 문법 | 의미 | 괄호 |
|------|------|------|------|
| 패키지 → 상수 | `math.pi` | math 안의 pi 값 | ❌ 없음 |
| 패키지 → 함수 | `math.sqrt(x)` | math 안의 sqrt 실행 | ✅ 있음 |
| 패키지 → 클래스 | `pd.DataFrame(...)` | pandas 안의 DataFrame 생성 | ✅ 있음 |
| 서브패키지 → 함수 | `np.random.randint()` | numpy.random 안의 함수 | ✅ 있음 |
| 클래스 내부 → 속성 저장 | `self.이름 = 값` | 이 인스턴스의 이름 속성에 저장 | ❌ 없음 |
| 클래스 내부 → 메서드 호출 | `self.메서드()` | 이 인스턴스의 메서드 실행 | ✅ 있음 |
| 인스턴스 → 속성 읽기 | `df.shape` | df의 shape 값 | ❌ 없음 |
| 인스턴스 → 메서드 실행 | `df.head()` | df의 head 메서드 실행 | ✅ 있음 |
| 체이닝 | `df.A().B().C()` | 순서대로 실행, 각 단계가 객체 반환 | ✅ 있음 |
