## 1장. 리스트 (List)

리스트는 여러 값을 **순서대로** 담는 자료구조입니다.  
pandas에서 DataFrame의 각 **컬럼은 내부적으로 리스트(정확히는 ndarray)**입니다.

```
상품명 컬럼  → ['노트북', '마우스', '키보드', ...]
가격 컬럼   → [1200000, 35000, 75000, ...]
```

````python
# =============================================
# 1-1. 리스트 기초: 생성과 접근
# =============================================

# 설명: 상품 판매 데이터를 리스트로 표현
# pandas 연결: 이 리스트들이 나중에 DataFrame의 각 컬럼이 됩니다
상품명 = ['노트북', '마우스', '키보드', '모니터', '헤드셋', '웹캠']
가격   = [1200000, 35000, 75000, 450000, 89000, 120000]
수량   = [5, 50, 30, 8, 20, 12]

# 설명: 인덱스로 특정 요소 접근 (0번째부터 시작)
# 주의: Python은 0-indexed! 첫 번째 요소는 [0]
print('첫 번째 상품:', 상품명[0])   # → 노트북
print('두 번째 가격:', 가격[1])     # → 35000

# 설명: 음수 인덱스로 뒤에서부터 접근
print('마지막 상품:', 상품명[-1])   # → 웹캠
print('뒤에서 두 번째:', 상품명[-2])  # → 헤드셋
````

````python
# =============================================
# 1-2. 리스트 슬라이싱: [시작:끝:간격]
# =============================================

# 설명: 리스트의 일부분만 잘라내는 슬라이싱
# pandas 연결: df.iloc[0:3]처럼 행/열을 잘라낼 때 같은 원리 사용

print('처음 3개 상품:', 상품명[0:3])    # → ['노트북', '마우스', '키보드']
print('2번째부터 끝까지:', 가격[2:])    # → [75000, 450000, 89000, 120000]
print('짝수 인덱스만:', 상품명[::2])    # → ['노트북', '키보드', '헤드셋']
print('역순 정렬:', 가격[::-1])         # → 리스트를 뒤집음
````

````python
# =============================================
# 1-3. 리스트 주요 메서드
# =============================================

# 설명: 리스트를 수정하는 다양한 메서드들
# pandas 연결: pandas의 열 추가/삭제도 내부적으로 유사한 동작

# append: 끝에 요소 추가
상품명.append('스피커')
가격.append(65000)
수량.append(25)
print('추가 후:', 상품명)  # 스피커가 추가됨

# len: 리스트 길이
print('상품 개수:', len(상품명))  # → 7

# sorted: 정렬된 새 리스트 반환 (원본 유지)
정렬된_가격 = sorted(가격)
print('가격 오름차순:', 정렬된_가격)
print('원본 가격 유지:', 가격)  # 원본은 그대로

# sum, min, max: 집계 함수
print(f'총 가격 합계: {sum(가격):,}원')
print(f'최소 가격: {min(가격):,}원')
print(f'최대 가격: {max(가격):,}원')
````

````python
# =============================================
# 1-4. 중첩 리스트 (2차원 리스트)
# =============================================

# 설명: 리스트 안에 리스트를 넣으면 표(행렬) 구조를 만들 수 있음
# pandas 연결: DataFrame은 본질적으로 2차원 데이터 구조
#              행(row) × 열(column)의 표 형태

# 각 행 = [상품명, 가격, 수량]
판매_데이터 = [
    ['노트북',  1200000, 5],
    ['마우스',    35000, 50],
    ['키보드',    75000, 30],
    ['모니터',   450000, 8],
    ['헤드셋',    89000, 20],
]

# 결과: 2차원 인덱싱으로 특정 셀 접근
print('2번째 행:', 판매_데이터[1])          # → ['마우스', 35000, 50]
print('3번째 행의 가격:', 판매_데이터[2][1])  # → 75000

# 설명: 이것이 바로 DataFrame의 구조와 같습니다!
# pandas에서는 판매_데이터[2][1] 대신 df.iloc[2, 1]로 접근
````

## 2장. 딕셔너리 (Dictionary)

딕셔너리는 **키(key): 값(value)** 쌍으로 데이터를 저장합니다.  

> **가장 중요한 pandas 연결**:  
> `pd.DataFrame(딕셔너리)` → 딕셔너리의 **키**가 컬럼명이 되고, **값(리스트)**이 컬럼 데이터가 됩니다!

```python
# 딕셔너리 구조
{'컬럼명1': [값1, 값2, ...], '컬럼명2': [값1, 값2, ...]}
        ↓ pd.DataFrame() 변환
# DataFrame 구조
   컬럼명1  컬럼명2
0    값1      값1
1    값2      값2
```

````python
# =============================================
# 2-1. 딕셔너리 기초: 생성과 접근
# =============================================

# 설명: 하나의 상품 정보를 딕셔너리로 표현
# pandas 연결: DataFrame의 한 행(row)도 딕셔너리처럼 접근 가능
상품_정보 = {
    '상품명': '노트북',
    '카테고리': 'IT기기',
    '가격': 1200000,
    '수량': 5,
    '평점': 4.8
}

# 키로 값에 접근
print('상품명:', 상품_정보['상품명'])    # → 노트북
print('가격:', 상품_정보['가격'])        # → 1200000

# get() 메서드: 키가 없어도 오류 없이 기본값 반환
# 주의: 없는 키를 직접 접근하면 KeyError 발생!
print('할인율:', 상품_정보.get('할인율', 0))  # → 0 (없으므로 기본값)

# 키 목록, 값 목록 확인
print('키 목록:', list(상품_정보.keys()))
print('값 목록:', list(상품_정보.values()))
````

````python
# =============================================
# 2-2. 딕셔너리 → pandas DataFrame 변환
# =============================================

import pandas as pd

# 설명: 컬럼별로 리스트를 담은 딕셔너리
# pandas 연결: 이것이 DataFrame을 만드는 가장 기본적인 방법!
판매_딕셔너리 = {
    '상품명':   ['노트북', '마우스', '키보드', '모니터', '헤드셋', '웹캠'],
    '카테고리': ['IT기기', 'IT기기', 'IT기기', 'IT기기', '음향기기', '영상기기'],
    '가격':     [1200000, 35000, 75000, 450000, 89000, 120000],
    '수량':     [5, 50, 30, 8, 20, 12],
    '지역':     ['서울', '부산', '서울', '대구', '서울', '인천'],
    '평점':     [4.8, 4.5, 4.7, 4.6, 4.3, 4.9],
}

# 딕셔너리를 DataFrame으로 변환
df = pd.DataFrame(판매_딕셔너리)

# 결과: 키(컬럼명)와 값(리스트)이 표 형태로 변환됨
print('DataFrame 생성 완료!')
print(df)
````

````python
# =============================================
# 2-3. DataFrame의 컬럼 접근 = 딕셔너리와 동일
# =============================================

# 설명: DataFrame 컬럼 접근 방법이 딕셔너리와 완전히 동일!
# pandas 연결: df['컬럼명'] ← 딕셔너리의 dict['키']와 동일한 문법

print('=== 딕셔너리 접근 ===')  
print(판매_딕셔너리['가격'])    # 딕셔너리에서 가격 리스트 가져오기

print('\n=== DataFrame 접근 ===' )
print(df['가격'])               # DataFrame에서 가격 컬럼 가져오기

# 결과: 둘 다 가격 데이터를 반환 (단, df는 Series 타입)
print('\n딕셔너리 타입:', type(판매_딕셔너리['가격']))
print('DataFrame 컬럼 타입:', type(df['가격']))  # → pandas Series
````

````python
# =============================================
# 2-4. 딕셔너리 수정과 DataFrame 컬럼 추가
# =============================================

# 설명: 딕셔너리에 새 키-값 쌍 추가
상품_정보['재고상태'] = '재고있음'
print('딕셔너리에 추가:', 상품_정보)

# pandas 연결: DataFrame에 새 컬럼 추가도 동일한 문법!
# 매출 = 가격 × 수량 계산하여 새 컬럼으로 추가
df['매출'] = df['가격'] * df['수량']
print('\n매출 컬럼 추가 후:')
print(df[['상품명', '가격', '수량', '매출']])
````

## 3장. for 루프

for 루프는 리스트나 딕셔너리의 요소를 **하나씩 순회**합니다.  

> **pandas 연결**:  
> `df.iterrows()` — DataFrame의 각 행을 for 루프로 순회  
> `df.itertuples()` — 더 빠른 행 순회 방법 (심화 노트북에서 자세히)

````python
# =============================================
# 3-1. 기본 for 루프
# =============================================

# 설명: 리스트의 모든 요소를 하나씩 꺼내어 처리
# pandas 연결: for 루프는 df.iterrows()의 기반 개념

상품_리스트 = ['노트북', '마우스', '키보드', '모니터', '헤드셋']

print('=== 상품 목록 출력 ===')
for 상품 in 상품_리스트:
    print(f'  - {상품}')  # f-string으로 변수를 문자열에 삽입

# 설명: enumerate()로 인덱스와 값을 동시에 가져오기
# pandas 연결: df.iterrows()도 (인덱스, 행데이터) 튜플을 반환
print('\n=== 번호 포함 출력 ===')
for 번호, 상품 in enumerate(상품_리스트, start=1):
    print(f'  {번호}번: {상품}')
````

````python
# =============================================
# 3-2. for 루프로 딕셔너리 순회
# =============================================

# 설명: 딕셔너리를 for 루프로 순회하는 3가지 방법
# pandas 연결: df.items()도 (컬럼명, Series) 쌍을 반환

상품_가격 = {'노트북': 1200000, '마우스': 35000, '키보드': 75000}

# 키만 순회
print('=== 키(상품명)만 순회 ===')
for 상품명 in 상품_가격:
    print(f'  {상품명}')

# 키-값 쌍 순회 (가장 자주 사용)
print('\n=== 키-값 쌍 순회 ===')
for 상품명, 가격 in 상품_가격.items():
    print(f'  {상품명}: {가격:,}원')

# pandas 연결: DataFrame도 동일한 방식으로 컬럼 순회 가능
print('\n=== pandas DataFrame 컬럼 순회 ===')
for 컬럼명, 컬럼데이터 in df[['상품명', '가격']].items():
    print(f'  컬럼 "{컬럼명}"의 첫 번째 값: {컬럼데이터.iloc[0]}')
````

````python
# =============================================
# 3-3. for 루프로 누적 계산
# =============================================

# 설명: for 루프를 이용한 합계/최대값 계산
# pandas 연결: df['가격'].sum(), df['가격'].max() 같은 집계 함수의 원리

가격_리스트 = [1200000, 35000, 75000, 450000, 89000, 120000]

# 합계 계산
총합 = 0
for 가격 in 가격_리스트:
    총합 = 총합 + 가격  # 총합 += 가격 으로도 쓸 수 있음
print(f'for 루프 합계: {총합:,}원')

# pandas는 이것을 한 줄로!
print(f'pandas 합계:   {df["가격"].sum():,}원')

# 최대값 찾기
최대값 = 가격_리스트[0]
for 가격 in 가격_리스트:
    if 가격 > 최대값:
        최대값 = 가격
print(f'\nfor 루프 최대값: {최대값:,}원')
print(f'pandas 최대값:   {df["가격"].max():,}원')
````

````python
# =============================================
# 3-4. pandas DataFrame 행 순회 (iterrows)
# =============================================

# 설명: df.iterrows()는 (인덱스, 행) 튜플을 반환
# 주의: iterrows()는 느립니다. 대용량 데이터는 벡터화 연산 사용 권장
#       하지만 for 루프 개념을 이해하면 iterrows()도 쉽게 이해됩니다

print('=== df.iterrows()로 행 순회 ===')
for 인덱스, 행 in df.iterrows():
    # 행은 딕셔너리처럼 컬럼명으로 접근 가능
    상품명 = 행['상품명']
    매출 = 행['매출']
    print(f'  [{인덱스}] {상품명}: 매출 {매출:,}원')
````

## 4장. 조건문 (if / elif / else)

조건문은 특정 조건에 따라 다른 코드를 실행합니다.  

> **pandas 연결**:  
> `df[df['가격'] > 100000]` — 조건에 맞는 행만 필터링  
> `df['등급'] = df['가격'].apply(lambda x: '고가' if x > 100000 else '저가')`

````python
# =============================================
# 4-1. 기본 if / elif / else
# =============================================

# 설명: 가격에 따라 상품 등급을 분류하는 함수
# pandas 연결: apply()와 함께 쓰면 전체 컬럼에 조건 적용 가능

def 가격_등급(가격):
    """가격에 따라 등급을 반환하는 함수"""
    if 가격 >= 500000:
        return '프리미엄'     # 50만원 이상
    elif 가격 >= 100000:
        return '중고가'       # 10만원 이상 50만원 미만
    elif 가격 >= 50000:
        return '중가'         # 5만원 이상 10만원 미만
    else:
        return '저가'         # 5만원 미만

# 개별 테스트
테스트_가격들 = [1200000, 450000, 89000, 75000, 35000]
for 가격 in 테스트_가격들:
    등급 = 가격_등급(가격)
    print(f'  가격 {가격:>10,}원 → {등급}')
````

````python
# =============================================
# 4-2. boolean 리스트와 pandas 필터링의 연결
# =============================================

# 설명: 조건문은 True/False 값을 만들고, 이를 필터링에 사용
# pandas 연결: df['가격'] > 100000 → boolean Series
#              df[boolean_Series] → 조건에 맞는 행만 추출

가격들 = [1200000, 35000, 75000, 450000, 89000, 120000]

# 일반 Python: 조건에 맞는 인덱스를 True/False로
고가_여부 = [가격 > 100000 for 가격 in 가격들]
print('고가 여부 (True/False):', 고가_여부)

# pandas에서도 동일한 원리!
print('\n=== pandas boolean 필터링 ===')
고가_조건 = df['가격'] > 100000          # boolean Series 생성
print('boolean Series:')
print(고가_조건.values)                   # [True, False, False, True, False, True]

print('\n고가 상품만 필터링:')
고가_상품 = df[고가_조건]                 # boolean으로 필터링
print(고가_상품[['상품명', '가격']])
````

````python
# =============================================
# 4-3. 복합 조건 (and, or, not)
# =============================================

# 설명: 여러 조건을 조합하는 방법
# 주의: pandas에서는 and/or 대신 &/| 기호 사용!

# 일반 Python: and/or 사용
가격 = 89000
평점 = 4.3

if 가격 < 100000 and 평점 >= 4.0:
    print(f'가격({가격:,}원)이 합리적이고 평점({평점})도 좋음!')

# pandas: & (and), | (or), ~ (not) 사용
print('\n=== pandas 복합 조건 필터링 ===')

# 가격 10만원 미만 AND 평점 4.5 이상인 상품
# 주의: 각 조건은 반드시 괄호로 묶어야 함!
조건 = (df['가격'] < 100000) & (df['평점'] >= 4.5)
결과 = df[조건]
print('가격 10만원 미만 & 평점 4.5 이상:')
print(결과[['상품명', '가격', '평점']])

# 서울 또는 인천 지역 상품
수도권_조건 = (df['지역'] == '서울') | (df['지역'] == '인천')
print('\n수도권(서울/인천) 상품:')
print(df[수도권_조건][['상품명', '지역']])
````

````python
# =============================================
# 4-4. for 루프 + 조건문 조합
# =============================================

# 설명: for 루프와 조건문을 함께 사용하여 데이터 분류
# pandas 연결: 이 패턴이 apply() + lambda의 원리

가격_리스트 = [1200000, 35000, 75000, 450000, 89000, 120000]
상품_리스트 = ['노트북', '마우스', '키보드', '모니터', '헤드셋', '웹캠']

print('=== 상품 등급 분류 (for + if) ===')
for 상품, 가격 in zip(상품_리스트, 가격_리스트):
    등급 = 가격_등급(가격)  # 앞서 정의한 함수 재사용
    print(f'  {상품:<8} | {가격:>10,}원 | {등급}')

# pandas 방식: apply()로 한 번에 처리 (훨씬 간결!)
print('\n=== pandas apply()로 동일 작업 ===')
df['가격등급'] = df['가격'].apply(가격_등급)
print(df[['상품명', '가격', '가격등급']])
````

## 5장. 종합 실습

지금까지 배운 **리스트, 딕셔너리, for 루프, 조건문**을 모두 활용합니다.

````python
# =============================================
# 5-1. 종합 실습: 판매 데이터 분석
# =============================================

# 설명: 지금까지 배운 개념을 모두 사용하는 종합 예제
# 목표: 카테고리별 매출 합계를 딕셔너리로 집계

# 데이터 준비 (딕셔너리 → DataFrame)
판매_데이터 = {
    '상품명':   ['노트북', '마우스', '키보드', '모니터', '헤드셋', '웹캠'],
    '카테고리': ['IT기기', 'IT기기', 'IT기기', 'IT기기', '음향기기', '영상기기'],
    '가격':     [1200000, 35000, 75000, 450000, 89000, 120000],
    '수량':     [5, 50, 30, 8, 20, 12],
}

# 1단계: 매출 계산 (가격 × 수량)
매출_리스트 = []
for 가격, 수량 in zip(판매_데이터['가격'], 판매_데이터['수량']):
    매출_리스트.append(가격 * 수량)
판매_데이터['매출'] = 매출_리스트

# 2단계: 카테고리별 매출 합계 (딕셔너리에 누적)
카테고리별_매출 = {}  # 빈 딕셔너리로 시작
for i in range(len(판매_데이터['상품명'])):
    카테고리 = 판매_데이터['카테고리'][i]
    매출 = 판매_데이터['매출'][i]
    
    if 카테고리 in 카테고리별_매출:  # 이미 있는 카테고리면 더하기
        카테고리별_매출[카테고리] += 매출
    else:                           # 새 카테고리면 추가
        카테고리별_매출[카테고리] = 매출

# 결과 출력
print('=== 카테고리별 매출 (Python 방식) ===')
for 카테고리, 매출 in 카테고리별_매출.items():
    print(f'  {카테고리}: {매출:,}원')

# pandas 방식: 단 두 줄로 동일 작업!
df['매출'] = df['가격'] * df['수량']
print('\n=== 카테고리별 매출 (pandas 방식) ===')
print(df.groupby('카테고리')['매출'].sum())
````
